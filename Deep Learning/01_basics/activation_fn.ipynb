Lets look at some Activation functions
We have different types of activation functions namely sigmoid, softmax RELU,tan h, LeakyRelu, PRelu, ELU. Among these sigmoid and softmax are used typically for the output layer, and other in DEnse layer.
Sigmoid for binary classification and softmax for multi class classification.
Sigmoid and tanh has vanishing gradient problem and makes learning process slow.


```python
import math
def sigmoid(x):
    return 1/(1+ math.exp(-x))
sigmoid(6)
```

```python
def tanh(x):
    return (2/(1+math.exp(-2*x)))-1
tanh(1)
```

```python
def relu(x):
    return max(0,x)
relu(3)
```

```python
relu(-9)
```

```python
def leaky_relu1(x):
    return max(0.1*x,x)
```

```python
leaky_relu1(6)
```
